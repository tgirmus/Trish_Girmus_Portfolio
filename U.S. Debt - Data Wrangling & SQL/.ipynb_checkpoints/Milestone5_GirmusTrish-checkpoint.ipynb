```python
# Importing third database from API/from Milestone 4
# Connecting to API for Interest Expense on the Debt Outstanding dataset
# import requests library
import requests

# URL for API: "https://api.fiscaldata.treasury.gov/services/api/fiscal_service/v2/accounting/od/interest_expense" 
# Assigning the home page URL to variable named url
url = "https://api.fiscaldata.treasury.gov/services/api/fiscal_service/v2/accounting/od/interest_expense?sort=-record_date&format=json&page[number]=1&page[size]=2242"

# Use get method from requests library to get response from this page
response = requests.get(url)
```

```python
# Import JSON library to read raw data 
import json
json_output = json.loads(response.text)
# print(json_output)
# print(type(json_output))
```

```python
metadata = json_output['meta']
```

```python
interest_expense_data = json_output['data']
```

```python
print(interest_expense_data)
```

```python
column_headers = list(interest_expense_data[0].keys())
```

```python
interest_expense_data_values = []

for i in range(0, len(interest_expense_data)):
    interest_expense_data_values.append(list(interest_expense_data[i].values()))
```

```python
import pandas as pd

# making a dataframe 
# 1st transformation: changing column headers/variable names so they are all uniform
expense_data_df = pd.DataFrame(data=interest_expense_data_values, columns=column_headers)
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://')
```

```python
# Converting debt_penny into SQL
debt_penny.to_sql('debt_penny', con = engine)
```

```python
engine.execute("SELECT * FROM debt_penny").fetchall()
```

```python
# Converting op_cashbal into SQL
op_cashbal.to_sql('op_cashbal', con = engine)
```

```python
engine.execute("SELECT * FROM op_cashbal").fetchall()
```

```python
# Converting expense_data_df into SQL
expense_data_df.to_sql('expense_data_df', con = engine)
```

```python
engine.execute("SELECT * FROM expense_data_df").fetchall()
```

```python
engine.execute('PRAGMA table_info(debt_penny)').fetchall()
```

```python
full_table = engine.execute('SELECT * FROM debt_penny JOIN op_cashbal ON debt_penny.record_date=op_cashbal.record_date JOIN expense_data_df ON expense_data_df.record_date=op_cashbal.record_date').fetchall()
# print(full_table)
```

```python
#engine.execute('CREATE TABLE full_table AS SELECT * FROM debt_penny JOIN op_cashbal ON debt_penny.record_date=op_cashbal.record_date JOIN expense_data_df ON expense_data_df.record_date=op_cashbal.record_date')
# table already exists which is why code is commented out - uncomment to rerun
```

```python
full_table_column_output = engine.execute('PRAGMA table_info(full_table)').fetchall()
column_names = [x[1] for x in full_table_column_output]
print(column_names)
```

```python
full_table_df = pd.DataFrame(data = full_table, columns = column_names)
print(full_table_df) # dataset is now 4,674 rows and 43 columns with combined 3 tables
```

```python
# Plot 3: Pie Chart - Fiscal YTD Expense Amount per Fiscal Year
import matplotlib.pyplot as plt

sum_2016 = sum(full_table_df[full_table_df['record_fiscal_year'] == '2016']['fytd_expense_amt'])
sum_2017 = sum(full_table_df[full_table_df['record_fiscal_year'] == '2017']['fytd_expense_amt'])
sum_2018 = sum(full_table_df[full_table_df['record_fiscal_year'] == '2018']['fytd_expense_amt'])
sum_2019 = sum(full_table_df[full_table_df['record_fiscal_year'] == '2019']['fytd_expense_amt'])
sum_2020 = sum(full_table_df[full_table_df['record_fiscal_year'] == '2020']['fytd_expense_amt'])
#print(sum_2016, sum_2017, sum_2018, sum_2019, sum_2020)
#print(full_table_df[full_table_df['record_fiscal_year'] == '2016'])
#df = pd.DataFrame(data={'amounts': [sum_2016, sum_2017, sum_2018, sum_2019, sum_2020], 'year': [2016, 2017, 2018, 2019, 2020]})
#df['amounts'].plot.pie()

labels = ['2016', '2017', '2018', '2019', '2020']
explode = (0, 0, 0, 0, 0.1)
plt.pie([sum_2016, sum_2017, sum_2018, sum_2019, sum_2020], labels = labels, autopct = '%1.1f%%', 
       shadow = True, startangle = 90)

plt.title("Fiscal YTD Expense Amount Per Fiscal Year", fontsize='20')

plt.show()
```

```python
# Plot 4: Line Plot - Total Public Debt Per Fiscal Year
full_table_df.plot.line(x="fiscal_yr", y="total_public_debt")
plt.title('Total Public Debt Per Fiscal Year')
plt.show()
```
